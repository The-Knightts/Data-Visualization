```python
import pandas as pd
```

```python
#Using the cleaned dataset created in the previous assignment 

dataset = pd.read_csv('C:\\Users\\KUBER\\OneDrive\\Documents\\ITR\\Py ITR\\Assignments\\Assignment-7\\clean_country_wise_data.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
dataset.columns
```

```python
TotalCases = dataset[['Confirmed', 'Deaths', 'Recovered']].cumsum()
    
plt.figure(figsize=(10, 6))
plt.plot(TotalCases.index, TotalCases['Confirmed'], label='Confirmed Cases')
plt.plot(TotalCases.index, TotalCases['Deaths'], label='Deaths')
plt.plot(TotalCases.index, TotalCases['Recovered'], label='Recovered')
plt.xlabel('Index')
plt.ylabel('Number of Cases')
plt.title('Total Confirmed, Deaths, and Recovered Cases')
plt.legend()
plt.show()
```

```python
TotalCases = dataset[['Confirmed', 'Deaths', 'Recovered']].cumsum()
    
plt.figure(figsize=(10, 6))
plt.scatter(TotalCases.index, TotalCases['Confirmed'], label='Confirmed Cases')
plt.scatter(TotalCases.index, TotalCases['Deaths'], label='Deaths')
plt.scatter(TotalCases.index, TotalCases['Recovered'], label='Recovered')
plt.xlabel('Index')
plt.ylabel('Number of Cases')
plt.title('Total Confirmed, Deaths, and Recovered Cases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
top_countries = dataset.groupby('Country/Region')['Confirmed'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
plt.bar(top_countries.index, top_countries)
plt.xlabel('Country/Region')
plt.ylabel('Number of Cases')
plt.title('Top 10 Countries with Highest Number of Cases')
plt.xticks(rotation = 90)
plt.show()
```

```python
top_countries = dataset.groupby('Country/Region')['Confirmed'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
plt.hist(top_countries, bins = 20)
plt.xlabel('Country/Region')
plt.ylabel('Number of Cases')
plt.title('Top 10 Countries with Highest Number of Cases')
plt.xticks(rotation = 90)
plt.show()
```

```python
daily_cases = dataset[['New cases', 'New deaths', 'New recovered']]
    
plt.figure(figsize=(10, 6))
plt.plot(daily_cases.index, daily_cases['New cases'], label='Daily New Cases')
plt.plot(daily_cases.index, daily_cases['New deaths'], label='Daily New Deaths')
plt.plot(daily_cases.index, daily_cases['New recovered'], label='Daily New Recovered')
plt.xlabel('Index')
plt.ylabel('Number of Cases')
plt.title('Daily New Cases, Deaths, and Recovered')
plt.legend()
plt.show()
```
